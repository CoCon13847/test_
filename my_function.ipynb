```python
def single_char_input():
    """
    사용자로부터 내 이름과 상대방 이름의 자음/모음을 한 글자씩 직접 입력받아
    궁합 점수를 계산합니다.

    Returns:
        tuple: (궁합 점수 (int), 궁합 메시지 (str))
    """

    # 각 자음과 모음에 부여할 숫자 딕셔너리
    # 이 숫자는 임의로 부여된 것이며, 필요에 따라 조정할 수 있습니다.
    char_scores = {
        'ㄱ': 1, 'ㄴ': 2, 'ㄷ':3, 'ㄹ':1,'ㅁ':3,'ㅂ':2,'ㅅ':4,'ㅇ':4,'ㅈ':4,'ㅊ':4,
        'ㅋ':2,'ㅌ':4,'ㅍ':4,'ㅎ':4,'ㅏ':3,'ㅑ':3,'ㅓ':2,'ㅕ':2,'ㅗ':3,'ㅛ':2,'ㅜ':2,'ㅠ':2,
        '-':1,'ㅣ':1,'ㅐ':4,'ㅖ':4,'ㅢ':4,'ㅔ':2

    }

    print("--- 궁합 계산기 (자모음 한 글자씩 입력) ---")
    print("예시: '이가은' -> ㅇ 엔터, ㅣ 엔터, ㄱ 엔터, ㅏ 엔터, ㅇ 엔터, ㅡ 엔터, ㄴ 엔터")
    print("입력을 마치려면 '끝'을 입력하고 엔터를 누르세요.")



    my_chars_list = []
    print("\n[내 이름의 자음/모음을 한 글자씩 입력하세요]")
    while True:
        char_input = input(f"내 이름의 {len(my_chars_list) + 1}번째 자음/모음 (종료: 끝): ").strip()
        if char_input == '끝':
            break
        if len(char_input) == 1 and char_input in char_scores:
            my_chars_list.append(char_input)
        else:
            print("❌ 올바른 한 글자 자음/모음이 아니거나 등록되지 않은 글자입니다. 다시 입력해주세요.")

    partner_chars_list = []
    print("\n[상대방 이름의 자음/모음을 한 글자씩 입력하세요]")
    while True:
        char_input = input(f"상대방 이름의 {len(partner_chars_list) + 1}번째 자음/모음 (종료: 끝): ").strip()
        if char_input == '끝':
            break
        if len(char_input) == 1 and char_input in char_scores:
            partner_chars_list.append(char_input)
        else:
            print("❌ 올바른 한 글자 자음/모음이 아니거나 등록되지 않은 글자입니다. 다시 입력해주세요.")




    # 입력된 자모음 리스트의 점수를 합산
    total_my_score = sum(char_scores.get(char, 0) for char in my_chars_list)
    total_partner_score = sum(char_scores.get(char, 0) for char in partner_chars_list)

    # 두 점수의 차이를 이용해 궁합 점수 계산
    # 점수 차이가 작을수록 궁합이 좋다고 가정합니다.
    max_possible_score = max(total_my_score, total_partner_score) * 2 # 대략적인 최대 점수 (과도하게 낮으면 0점 방지)
    
    if max_possible_score == 0 and (len(my_chars_list) > 0 or len(partner_chars_list) > 0): 
        # 점수가 0이지만 입력은 받은 경우 (등록되지 않은 글자만 입력된 경우)
        affinity_score = 0
    elif max_possible_score == 0: # 둘 다 입력이 아예 없는 경우
        affinity_score = 0 # 또는 다른 기본값
    else:
        difference = abs(total_my_score - total_partner_score)
        affinity_score = max(0, 100 - int((difference / max_possible_score) * 100))


    # 궁합 점수에 따른 메시지
    if affinity_score >= 90:
        message = "와우! 환상의 궁합이네요! 💯"
    elif affinity_score >= 70:
        message = "꽤 좋은 궁합입니다! 😊"
    elif affinity_score >= 50:
        message = "나쁘지 않은 궁합이에요. 노력하면 더 좋아질 거예요! 😉"
    else:
        message = "음... 좀 더 알아가 봐야 할 것 같아요. 😅"

    return affinity_score, message

# --- 함수 실행 ---
score, msg = single_char_input()

print(f"\n✨ 궁합 결과는...")
print(f"💖 궁합 점수: {score}점")
print(f"💬 메시지: {msg}")
```
